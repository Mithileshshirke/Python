```python
#handling outliers
#we will not look for outliers because we have single observations for each model 
```

```python
#categorical data type
#divide data set into categorical and numerical type
df1=df.select_dtypes(object)
df2=df.select_dtypes(['float','int64'])
```

```python
print('Categorical data :')
print(df1.info())
print('Numerical data :')
print(df2.info())
```

```python
#transform categorical data to numerical using labelEncoder
from sklearn.preprocessing import LabelEncoder
for col in df1:
    obj=LabelEncoder() #create object of class
    df1[col]=obj.fit_transform(df1[[col]])
df1.head()
```

```python
#handling skewness on df2 numerical data set
from scipy.stats import skew
print('Skewness')
for col in df2:
    print(col)
    print(skew(df2[col]))
    sns.distplot(df2[col])
    plt.show()
```

```python
#find correlation
plt.figure(figsize=(8,8))
sns.heatmap(df2.corr(),annot=True)
plt.show()
```

```python
#correlation between mpg and gear & mpg and qsec is not so good
#check whether gear or qsec column having negative values or not
df2[df2['gear']<0]
```

```python
df2[df2['qsec']<0]
```

```python
#so both columns are not having negative values and positive skewness and not good relation
#so we transform these columns by taking square root
```

```python
df2['qsec']=round(np.sqrt(df2['qsec']),1)
df2['gear']=round(np.sqrt(df2['gear']),1)
```

```python
df2.head()
```

```python
df_new=pd.concat([df1,df2],axis=1)
df_new.head()
```

```python
df_new.info()
```

```python
X=df_new.drop('mpg',axis=1)
Y=df_new['mpg'] #here mpg is target variable
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
for col in X:
    obj=StandardScaler()
    X[col]=obj.fit_transform(X[[col]])
```
