Q.what is filter in pandas write a program in python  pandas for filter

-Pandas dataframe.filter() function is used to Subset rows or columns of dataframe

according to labels in the specified index.

```python
import pandas as pd
df=pd.read_csv('automobile_data.csv')#read_csv() inbuilt method to import data from file
df.head()
```

```python
#subset rows or columns from df
df.filter(['make','fuel-type','price'])#filter() inbuilt method to subset data
```
