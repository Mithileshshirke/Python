Scrape IMDB movie rating and details using Python

```python
import requests
import bs4
```

```python
url='https://www.imdb.com/chart/top/'
response=requests.get(url)
soup=bs4.BeautifulSoup(response.text)
tables=soup.find_all('table')
table=tables[0]
rows=table.find_all('tr')
data=rows[1:]
for row in data:
    data1=row.find_all('td',{'class':"titleColumn"})
    data2=row.find_all('td',{'class':"ratingColumn imdbRating"})
    print('Rank & Title : ',data1[0].text)
    print('Rating : ',data2[0].text)
    
```
